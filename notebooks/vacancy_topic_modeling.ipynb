```python
from bertopic import BERTopic
```

```python
from sentence_transformers import SentenceTransformer

sentence_model = SentenceTransformer("distiluse-base-multilingual-cased-v1", device="cuda")
topic_model = BERTopic(embedding_model=sentence_model,verbose=True, n_gram_range=(1,2))
```

```python
 df_bd['Description'] =  df_bd['Description'].apply(lambda x: x.lower())
```

```python
from tqdm.contrib.concurrent import process_map  # or thread_map
```

```python
from nltk.corpus import stopwords
import re
def clean_dataset(senteces):
    text = senteces.lower()
    text = re.findall(r'\w+', text)
    tokenize = [i for i in text if not i.isdigit()]
    tokenize = [word for word in tokenize if not word in stopwords.words()]
    text = ' '.join(tokenize)
    return text
```

```python
desc_course = process_map(clean_dataset, list(df_bd['Description']), max_workers=7)
```

```python
docs = df_bd['Description']
```

```python
topics, probabilities = topic_model.fit_transform(docs)
```
