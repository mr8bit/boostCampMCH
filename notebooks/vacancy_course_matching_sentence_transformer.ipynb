```python
import pandas as pd
import glob
import numpy as np
data_hh_list = glob.glob("./data_courses/*.csv")
df_course = pd.DataFrame(columns=['name','img','desc','url'])

for csv in data_hh_list:
    df_course = df_course.append(pd.read_csv(csv))
df_course = df_course.drop(['Unnamed: 0'], axis=1)
df_course = df_course.dropna()
df_course = df_course.reset_index(drop=True)
len(df_course)
df_course = df_course.drop_duplicates(subset="name")
df_course = df_course.reset_index(drop=True)
len(df_course)
```

```python
from nltk.corpus import stopwords
import re
def clean_dataset(senteces):
    text = senteces.lower()
    text = re.findall(r'\w+', text)
    tokenize = [i for i in text if not i.isdigit()]
    tokenize = [word for word in tokenize if not word in stopwords.words()]
    text = ' '.join(tokenize)
    return text
```

```python
desc_course = process_map(clean_dataset, list(df_course['desc']), max_workers=6)
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('quora-distilbert-multilingual')
```

```python
embeddings_course = model.encode(desc_course)
```
